# Exploring Exonerations in the United States

## Introduction
-----

Our research explores exonerations in the United States from 1956 to 2021. There have been over 3,303 officially recorded exonerations—in which a person who was convicted of a crime is officially cleared—and doubtlessly many other innocent defendants who were never exonerated. We are interested in whether or not a relationship exists between the amount of time between conviction and exoneration (referred to in our report as "Exonerated Time Range") and various factors such as race, sex, and U.S. state. 

Given the context of race in U.S. history and politics, we seek to understand how the length of time before exoneration has changed over the years across the different regions of the country for different races. Data has shown that African Americans are more likely than white Americans to be arrested, and once arrested, they are more likely to be convicted (https://www.sentencingproject.org/reports/report-to-the-united-nations-on-racial-disparities-in-the-u-s-criminal-justice-system/). As race relations and legislation has continually changed since the 1950s, especially following the Civil Rights Movement of the 1960s, we seek to explore whether these societal and legal changes have affected false convictions over time. We believe this may vary in different regions of the U.S., some of which historically have more cases of racial discrimination and mistreatment of minority groups. In addition, differences in state laws may have an effect on how certain crimes are treated. Therefore, our primary overarching research question is as follows: In cases of exoneration, is there an association with various contributing factors, demographics, type of crime, and time (exoneration time range and year convicted)?

Our research centers around data from the National Registry of Exonerations, which tracks exonerations for convictions going back to 1956 and provides details of each case. Key variables that we use in our analysis include:

- **Exonerated Time Range:** The difference between the year of conviction and year of exoneration. In other words, how long it took for the convicted individual to be exonerated.
- **Race:** The race of the individual: Black, White, Hispanic, Asian, Native American, Other, or Don't Know.
- **Sex:** The sex of the individual: Male or Female.
- **State:** The state in which the individual was convicted.
- **Worst Crime:** The worst crime that the individual was convicted of.
- **DNA:** Indicates if post-conviction DNA testing was conducted and contributed to the exoneration.

The dataset and all variables are defined more thoroughly in the Data Description section below.

## Data Description

### Motivation
**Why was this dataset created?**

According to the mission statement of the project, the primary goal in creating this dataset, the National Registry of Exonerations, is to “reform the criminal justice system and reduce if not eliminate these tragic errors \[of wrongful convictions\] in the future. We also aim to make police officers, prosecutors, defense attorneys and judges more sensitive to the problem of wrongful convictions and more willing to reconsider the guilt of defendants who have already been convicted when new evidence of innocence comes to light” (https://www.law.umich.edu/special/exoneration/Pages/mission.aspx).

## Preregistration Statement

### Hypothesis 1:

**Research Question:** Is there a relationship between Exonerated Time Range (time between conviction year and exoneration year) and state, sex, race, worst crime, conviction year, and DNA? Are there interactions between race, sex, state, worst crime, conviction year, and DNA? 

**Hypothesis:** Exonerated time range increases with more severe worst crime, presence of DNA evidence in exoneration, and also increases if the exoneree's sex is male or if their race is Black, Hispanic, or Asian (i.e., not White). It decreases with conviction year (i.e. decreases as years increase). Exonerated Time range increases as the state becomes more southern in location.
($\beta_0, \beta_1, \beta_2, \beta_3, \beta_4, \beta_5, \beta_6,\beta_7,\beta_8 \neq 0$),
(Beta values in intended statistical analyses)

To answer this first research question, we will perform a multivariate linear regression with Exonerated Time range as the output and Sex, Race, Worst Crime, Conviction year, and DNA as the inputs. We will also include interaction effects between Race and Gender, State and Worst Crime, Conviction year and DNA. A visual representation is: 

Exonerated Time Range ~ Sex + Race + Worst Crime + Conviction Year + DNA + Race * Sex + State * Worst Crime + Conviction Year * DNA

The general fitted model is: 

Exonerated Time Range= $\beta_0 + \beta_1 Sex + \beta_2 Race + \beta_3 Worst Crime + \beta_4 Conviction Year + \beta_5 DNA + \beta_6 Race * Sex + \beta_7 State * Worst Crime + \beta_8 Convinction Year * DNA$

### Hypothesis 2:
**Research Question:**  Given the conviction year and a contributing factor of false conviction (e.g. Mistaken Witness Identification [MWID]), can we find the probability that the exoneree is each given race?

**Hypothesis:** The probability that an exoneree belongs to a minority racial group decreases with conviction year (i.e. decreases as years increase) and increases if Mistaken Witness Identification was a contributing factor of the false conviction.

($\beta_1$ of black, $\beta_1$ of hispanic, $\beta_1$ of asian, $\beta_1$ of native american > $\beta_1$ of white)

($\beta_2$ of black, $\beta_2$ of hispanic, $\beta_2$ of asian, $\beta_2$ of native american > $\beta_2$ of white)

We will check use a ztest in order to evaluate the statistical significance of coefficients of black, hispanic, asian,
native american to white. 

To answer this second research question, we will perform multiple multivariate logistic regression. For each logistic regression, we will code the outcome as a binary variable for whether or not the person belongs to the race group. The predictors will be Convicted year and Mistaken Witness Identification (MWID). A visual representation of the logisistic regressions are: 

Black ~ Convicted Year + MWID

White ~ Convicted Year + MWID

Hispanic ~ Convicted Year + MWID

Asian ~ Convicted Year + MWID

Native American ~ Convicted Year + MWID

The general fitted models are: 

logit($\pi$(black)) = $\beta_0 + \beta_1 Convicted Year + \beta_2 MWID$

logit($\pi$(white)) = $\beta_0 + \beta_1 Convicted Year + \beta_2 MWID$

logit($\pi$(hispanic)) = $\beta_0 + \beta_1 Convicted Year + \beta_2 MWID$

logit($\pi$(asian)) = $\beta_0 + \beta_1 Convicted Year + \beta_2 MWID$

Then, we utilized bootstrapping to find 95% confidence interval of the coefficients.

The histogram is the histogram of the coefficient for Alaska. This serves as a confidence check to see if the bootstrapping method was successful. 

The purpose of this bootstrapping method is to have more confidence that we reduced our model to include only significant factors of exonerated time range.

```python
#Bootstrapping: 
bs_beta = np.zeros(1000)
count_of_zeroes = np.zeros(130)
all_values = pd.DataFrame()
for i in range(1000):
    samples = h1_exon_data.sample(n=3191, replace=True)
    x = samples[dummies]
    y = samples['ExoneratedTimeRange']
    model = LinearRegression().fit(x, y)
    bs_beta[i] = model.coef_[0]
    all_values = pd.concat([all_values, pd.DataFrame(model.coef_).T], axis=0)
    
#can find mean + stdev
graph = seaborn.histplot(data=bs_beta)
```

```python
#Bootstrap analysis 
all_values.columns = dummies
all_ci = pd.DataFrame()
lower_bound = all_values.describe(include='all').loc['mean'] - 2*all_values.describe(include='all').loc['std']
upper_bound = all_values.describe(include='all').loc['mean'] + 2*all_values.describe(include='all').loc['std']

all_ci = all_ci.append(lower_bound, ignore_index=True)
all_ci = all_ci.append(upper_bound, ignore_index=True)
print(all_ci.T)
```

The next step step was to repeat the same process of linear regression and bootstrapping with our model that includes
interaction effects. 

```python
#Including interactions: interactions race * gender, state * worst crime, convictionyear * dna 
#race * gender
interactions = [] 
#error can be fixed by changing Other in Crime and Other etc. 
interact_exon_data = h1_exon_data

for i in sexes: 
    for y in races: 
        interact_col_name = i + "_" + y
        interactions.append(interact_col_name)
        h1_exon_data[interact_col_name] = h1_exon_data[i] * h1_exon_data[y]  

interact_exon_data["Convicted_DNA"] = interact_exon_data["Convicted"] * interact_exon_data["DNA"] 


for i in states: 
    temp = pd.DataFrame()
    for y in crimes:
        interact_col_name = i + "_" + y
        interactions.append(interact_col_name)
        temp[interact_col_name] = h1_exon_data[i] * h1_exon_data[y] 
    interact_exon_data = pd.concat([interact_exon_data,temp], axis=1)
```

```python

list_exon_data_col = interact_exon_data.columns.values.tolist()

dummies_interactions = list_exon_data_col[26:]
dummies_interactions.append('DNA')
dummies_interactions.append('Convicted')

betas = create_model(interact_exon_data[dummies_interactions], interact_exon_data['ExoneratedTimeRange'])
values = list(zip(dummies_interactions, betas.coef_))

# print(list_exon_data_col)
```

```python
#Reducing the Model 
#Remove Coeff 0: does not exist in our data 
print(interact_exon_data.shape)
drop_list = []
for i in interact_exon_data[dummies_interactions]:
    if interact_exon_data[i].sum(axis=0) == 0: 
        drop_list.append(i)
interact_exon_data = interact_exon_data.drop(columns = drop_list)
print(interact_exon_data.shape)
temp_list = interact_exon_data.columns[26:]


#Remove not significant values 
model = smf.OLS(exog=interact_exon_data[temp_list], endog= interact_exon_data['ExoneratedTimeRange'])
fit_model = model.fit()
interact_pval = fit_model.pvalues
temp_array = pd.DataFrame([temp_list, interact_pval, fit_model.params]).T
print(fit_model.summary())
temp_array =  temp_array.loc[(temp_array[1] <= 0.05)]
```

## Conclusion

The purpose of this project is to analyze if there an association with various contributing factors, demographics, type of crime, and time (exonerated time range and year convicted). From the first hypothesis, which was a multivariate linear regression, we wanted to analyze the relationship between exonerated time range and sex, race, worst crime, convicted year, presence of DNA evidence, and state where the crime was committed. We hypothesized that exonerated time range increases with more severe worst crime, presence of DNA evidence in exoneration, and also increases if the exoneree's sex is female or if their race is Black, Hispanic, or Asian (i.e., not White). It decreases with conviction year (i.e. decreases as years increase). Exonerated Time Range increases as the state becomes more southern in location.

The analysis of hypothesis 1 shows that the state where the crime was committed and the type of crime the exonerated individual committed are the most significant factors that influence exonerated time range. Using the bootstrapping methods, those factors were the only ones to come out statistically significant to predict exonerated time range. The analysis also shows that states on the east coast of the United States have longer exonerated time ranges. Types of crimes like murder, manslaughter, child sex abuse, and assault greatly increase the exonerated time range. One key observation from this analysis is that lower degrees of murder cases, like attempted murder and manslaughter, is associated with longer exonerated time range than higher degrees of murder cases. One reason this could be is because actual murder is harder to prove than attempted murder or manslaughter, which may lack concrete evidence. Therefore, individuals who are in prison for actual murder might be exonerated much quicker due to lack of evidence. 

Our hypothesis was partially rejected because a lot of the factors that we thought would affect exonerated time range were not significant. Our bootstrapping method was able to do feature selection and remove a lot of variables that included sex, race, presence of DNA evidence, and convicted year. However for the factors that were significant, our hypothesis was rejected. We hypothesized that exonerated time range increases as the state becomes more southern in loationl however, our analysis finds that exonerated time range increases when the crime in committed in the east coast of the United States. We also hypothesized that exonerated time range increases with more severe worst crime, but that wasn't the case. We found that in the cases of murder, attempted murder or manslaughter has lower exonerated time range than actual murder cases. 

In the second hypothesis, we wanted to analyze how the probability of an exonerated individual being in a certain race group is affected by presence of mistaken witness identification and convicted year. We hypothesized that the probability that an exoneree belongs to a minority racial group decreases with conviction year (i.e. decreases as years increase) and increases if Mistaken Witness Identification was a contributing factor of the false conviction. The first part of our analysis was supported by our hypothesis. However, we found that the probability of an exoneree belonging to any of the race groups decreases with conviction year. The second part of our hypothesis was partially supported. All minority groups except Asian people had increase in probability of an exoneree belonging to that group if there was a presence of mistaken witness identification. As hypothesized, people belonging to the white racial group has an increase in probability of belonging to that racial group if there was a presence of mistaken witness identification.

In hypothesis 2, for every one year increase in the convicted year, an exonerated individual has the lowest odds of being black, in comparison to the other race groups. This shows that as time has gone by, black people are less likely to be exonerated in comparison to other race groups. We also see that when there is a presence of mistaken witness identification, there is a decrease in the odds of an exonerated individual to be White and Asian; however, there is an increase in oddds of an individual to be Black, Hispanic, and Native American. This shows that Black, Hispanic, and Native American people are more likely to have mistaken witness identification in their criminal cases, which then affects their exonerated time range. 

This project shows that there are many factors to determine how long it takes an individual to be exonerated. This project highlights how different regions of the United States treat different types of crimes depending on the state laws. This project also highlights how belonging to a minority racial group has disadvantages when it comes to the U.S. justice system. Studies have been conducted that demonstrate the existence of "own-race" bias, meaning that people more frequently misidentify individuals of a different race from their own. 42% of wrongful convictions based on mistaken witness identity were cross-racial misidentifications (https://mtinnocenceproject.org/cross-racial-witness-misidentification/). Moving forward, there has to be more research that explores how to mitigate the issue of exonerees belonging to minority groups being more likely to experience mistaken witness identifiation. Countless innocent individuals could be saved from a wrongful conviction.

## Limitations

* The database does not include about 1,900 innocent defendants who were exonerated as a result of 15 large-scale police scandals because “they are the results of a collective exoneration based on problems in individual agencies” (https://www.huffpost.com/entry/national-registry-of-exonerations_n_1534030). This is a limitation because it is a very large number of individuals that, if included, would increase the size of the database by over 50% (it currently has 3,266 individuals). This would greatly increase the sample size of the data, and an increased sample size is more likely to be a more accurate representation of the actual population and thus lead to greater confidence and significance of results.


* The database focuses on exonerations because they are the only confirmed cases of false convictions. However, the true number of false convictions is undoubtedly higher. Similar to the previous limitation, this affects how accurately the data represents the true population. Specifically, when considering race as a factor, systemic racism in the judicial system likely has an effect on whether or not an individual is exonerated. According to the 2018 Report to the United Nations on Racial Disparities in the U.S. Criminal Justice System, African Americans are more likely than white Americans to be arrested, and once arrested, they are more likely to be convicted (https://www.sentencingproject.org/reports/report-to-the-united-nations-on-racial-disparities-in-the-u-s-criminal-justice-system/). It follows that this disproportion is likely to carry over into false convictions and exonerations as well. Therefore, the exonerated individuals included in the dataset may not be representative of all people who have been falsely convicted, particularly when it comes to race.


* In our data cleaning process, we eliminated any rows that included NaNs. The first limitation caused by this is the loss of data, which decreases our sample size. In addition, data is more likely to be missing from earlier years: records may not have been as well-kept as in the modern day, especially before digital records were introduced. By eliminating rows with NaNs, then, we are eliminating more data from earlier times. This may negatively affect our results when examining trends over time, since we have less data from the beginning of the time range.
